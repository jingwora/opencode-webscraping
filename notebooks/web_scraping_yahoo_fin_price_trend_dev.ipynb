```
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from pylab import rcParams
```

```
url = "https://query1.finance.yahoo.com/v8/finance/chart/COM7.BK?region=US&lang=en-US&includePrePost=false&interval=1wk&range=5y&corsDomain=finance.yahoo.com&.tsrc=finance"
```

```
response = requests.get(url)  # GET Method
```

```
response.status_code # 200 is OK, 404 is Not Found
```

```
js = response.json()
print(js)
```

```
result = js['chart'] ['result'] [0]
print(result)
```

```
meta = result['meta']
symbol = result['meta']['symbol']
exchangeName = result['meta']['exchangeName']
dataGranularity = result['meta']['dataGranularity']
range = result['meta']['range']
validRanges = result['meta']['validRanges']


print(meta)
print(symbol)
print(exchangeName)
print(dataGranularity)
print(range)
print(validRanges)
```

```
timestamp = result['timestamp']
open = result['indicators']['quote'][0]['open']
close = result['indicators']['quote'][0]['close']
high = result['indicators']['quote'][0]['high']
low = result['indicators']['quote'][0]['low']
volume = result['indicators']['quote'][0]['volume']

print(timestamp)
print(open)
print(close)
print(high)
print(low)
print(volume)
```

```
df = pd.DataFrame(list(zip(timestamp, open, close, high,  low, volume)), 
               columns =['timestamp', 'open', 'close', 'high', 'low', 'volume'])
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s') 
df = df.drop(columns=['timestamp'])
df['VWAP'] = ((df['volume'] * ((df['high'] + df['low'] + df['close'])/3)).cumsum())/df['volume'].cumsum()
df
```

```
df.describe(include = 'all')
```

```
# Create chart from plotly

chart_title = ('symbol: ' + symbol 
               + ' | market: ' + exchangeName 
               + ' | dataGranularity: ' + dataGranularity 
               + ' | range: ' + range)

layout =  go.Layout(plot_bgcolor='rgba(0,0,0,0)')

fig = go.Figure()
fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Scatter(
    x=df['datetime'], 
    y=df['close'], 
    name='close',
    line_color='yellow',
    mode='lines'), 
    row=1, col=1)
fig.add_trace(go.Scatter(
    x=df['datetime'], 
    y=df['VWAP'],
    name='VWAP',
    mode="lines"), 
    row=1, col=1)
fig.add_trace(go.Bar(
    x=df['datetime'].index, 
    y=df['volume'],
    name='volume'), 
    row=2, col=1)

fig.update_layout(title=chart_title,
                  width=1000, height=600, 
                  plot_bgcolor='rgb(0,0,0)')
fig.show()
```

```
# Create chart from matplotlib

rcParams['figure.figsize'] = 14,10
rcParams['axes.facecolor'] = 'black'
rcParams['legend.facecolor'] = 'white'

chart_title = ('symbol: ' + symbol 
               + ' | market: ' + exchangeName 
               + ' | dataGranularity: ' + dataGranularity 
               + ' | range: ' + range)

plt.subplot(2, 1, 1)
plt.title(chart_title)
plt.plot(df['datetime'], df['close'], label = 'close', color = 'yellow')
plt.plot(df['datetime'], df['VWAP'], label='VWAP', color='blue')
plt.legend()
plt.subplot(2, 1, 2)
plt.bar(df.index, df['volume'], label = 'volume', color = 'gray')
plt.legend()
plt.show()
```
